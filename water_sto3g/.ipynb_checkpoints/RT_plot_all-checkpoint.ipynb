```python
import h5py
import matplotlib.pyplot as plt
import numpy
from matplotlib.pyplot import figure

from functional import *
```

```python
# Set damping factor to get lifetime broadening
DAMP = 0.01
```

```python
# Pass in binary file name and the path to which the data are stored in
def read_dataset(fname, path):
    f = h5py.File(fname)
    return np.array(f[path])
```

```python
# Raw data from Chronus
time_c = read_dataset('water_x_0p0001.bin', "RT/TIME")*0.02418884254
ux_c = read_dataset('water_x_0p0001.bin', "RT/LEN_ELEC_DIPOLE")[:,0] / 0.393456
uy_c = read_dataset('water_y_0p0001.bin', "RT/LEN_ELEC_DIPOLE")[:,1] / 0.393456 + 0.00013
uz_c = read_dataset('water_z_0p0001.bin', "RT/LEN_ELEC_DIPOLE")[:,2] / 0.393456
dt_c = time_c[1] - time_c[0]
signals_c = [ux_c, uy_c, uz_c]
```

```python
# Raw data from Gaussian
x_data_g = np.genfromtxt("x_data_gaussian.txt", dtype=None)
y_data_g = np.genfromtxt("y_data_gaussian.txt", dtype=None)
z_data_g = np.genfromtxt("z_data_gaussian.txt", dtype=None)
time_g = x_data_g[1:, 0]*0.02418884254
dt_g = time_g[1] - time_g[0]
ux_g = x_data_g[1:, 2] 
uy_g = y_data_g[1:, 3]
uz_g = z_data_g[1:, 4]
signals_g = [ux_g, uy_g, uz_g]
```

```python
fig = plt.figure(figsize = (8, 6), dpi=300)
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

ax1.plot(time_c, ux_c, label= 'CQ')
ax1.plot(time_g, ux_g, label= 'Gaussian' )
ax1.legend(loc='upper left')

ax2.plot(time_c, uy_c, label= 'CQ')
ax2.plot(time_g, uy_g, label= 'Gaussian')
ax2.legend(loc='upper left')

ax3.plot(time_c, uz_c, label= 'CQ')
ax3.plot(time_g, uz_g, label= 'Gaussian')
ax3.legend(loc='upper left')

# Set common labels
fig.text(0.5, 0.01, 'Time (fs)', ha='center', va='center')
fig.text(0.01, 0.5, 'Dipole (Debye)', ha='center', va='center', rotation='vertical')

ax1.set_title('X direction')
ax2.set_title('Y direction')
ax3.set_title('Z direction')

fig.tight_layout() 


plt.show()
```

```python
fig, [ax1, ax2, ax3] = plt.subplots(3, 1, sharex=True)
ax1.plot(time_c, ux_c, label= 'CQ')
ax2.plot(time_c, uy_c, label= 'CQ')
ax3.plot(time_c, uz_c, label= 'CQ')
ax1.plot(time_g, ux_g, label= 'G' )
ax2.plot(time_g, uy_g, label= 'G')
ax3.plot(time_g, uz_g, label= 'G')
```

```python
# Fourier transformed spectra
spects_c = []
freq_c = None
for s in signals_c:
    s -= s[0]
    s *= np.exp(-DAMP*time_c)
    #w, f = fourier_tx(s, dt, 0)
    w, f = pade_tx(s, dt_c, (0,5))
    freq_c = w
    spects_c.append(f)
```

```python
# Fourier transformed spectra
spects_g = []
freq_g = None
for s in signals_g:
    s -= s[0]
    s *= np.exp(-DAMP*time_g)
    #w, f = fourier_tx(s, dt, 0)
    w, f = pade_tx(s, dt_g, (0,5))
    freq_g = w
    spects_g.append(f)
```

```python
total_c = sum([f.imag for f in spects_c]) / 3
total_c *= -freq_c

total_g = sum([f.imag for f in spects_g]) / 3
total_g *= -freq_g
```

```python
# Plotting
figure(figsize = (8, 6), dpi=300)
plt.plot(freq_c*27.2114, total_c, ls="-", linewidth = 2, label = "Chronus")
plt.plot(freq_g*27.2114, total_g, ls="--", linewidth = 2, label = "Gaussian")

#plt.ylim(0,0.175)
plt.ylabel("Intensity", fontsize=12)
plt.xlabel("Energy (eV)", fontsize=12)
plt.tight_layout()
plt.title("Water RT B3LYP Spectrum")
plt.legend(loc='upper left')
#plt.savefig("water_fourier_thin.pdf")
plt.show()
```
